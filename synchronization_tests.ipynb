Frequency Encoding

```
from collections import defaultdict
def frequency_encoding(data):
    encoding_dict = defaultdict(int)
    for row in data:
        encoding_dict[row[0]] += 1
    for row in data:
        row[0] = encoding_dict[row[0]]
    return data
```

Create Labels with 6 clusters

```
kmeans = KMeans(
   ...:     init="random",
   ...:     n_clusters=6,
   ...:     n_init=10,
   ...:     max_iter=300,
   ...:     random_state=42
   ...: )

kmeans.fit(x)

kmeans_list=[]

for i in range(len(kmeans.labels_)):
  kmeans_row=[]
  kmeans_row.append(kmeans.labels_[i]/10)
  kmeans_list.append(kmeans_row)

```

Add the clusters as a new feature to the train data

```
print(x[0])
x=np.concatenate((x, kmeans_list), axis=1)
print(x[0]);
sc = StandardScaler()

xx = sc.fit_transform(x)
data_train, data_test, labels_train, labels_test = train_test_split(xx, y, test_size=0.20, random_state=42)

```

Find the best tune for the algorithms by using gridsearch.

```
def display(results):
    print(f'Best parameters are: {results.best_params_}')
    print("\n")
    mean_score = results.cv_results_['mean_test_score']
    std_score = results.cv_results_['std_test_score']
    params = results.cv_results_['params']
    for mean,std,params in zip(mean_score,std_score,params):
        print(f'{round(mean,3)} + or -{round(std,3)} for the {params}')
```
